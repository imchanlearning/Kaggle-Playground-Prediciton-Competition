# Feature_eng

```python
def feat_eng(df):

    df['age_category'] = pd.cut(df['age'], bins=[0, 40, 60, 100], labels=['young', 'middle_age', 'senior'])

    df['bmi'] = df['weight(kg)'] / ((df['height(cm)'] / 100) ** 2)

    # Average visual and hearing score
    df['avg_eye_hear'] = (df['eyesight(left)'] + df['eyesight(right)']
                       + df['hearing(left)'] + df['hearing(right)']) / 4

    def cat_blood_pressure(systolic, cholesterol):
        if systolic < 120 and cholesterol < 200:
            return 'normal'
        elif 120 <= systolic < 140 or 200 <= cholesterol < 240:
            return 'prehypertensive'
        else:
            return 'hypertensive'
        
    df['blood_pressure_category'] = df.apply(lambda row: cat_blood_pressure(row['systolic'], row['Cholesterol']), axis=1)

    def cat_blood_sugar(fasting_blood_sugar):
        if fasting_blood_sugar < 100:
            return 'normal'
        elif 100 <= fasting_blood_sugar < 126:
            return 'prediabetic'
        else:
            return 'diabetic'
    df['blood_sugar_cat'] = df['fasting blood sugar'].apply(cat_blood_sugar)

    # Additional Ratios
    df["HDL-LDL Ratio"] = df["HDL"] / df["LDL"]
    df["HDL-triglyceride Ratio"] = df["HDL"] / df["triglyceride"]
    df["LDL-triglyceride Ratio"] = df["LDL"] / df["triglyceride"]

    return df
```

```python
df=feat_eng(df)
df = pd.get_dummies(df, columns=['age_category', 'blood_pressure_category', 'blood_sugar_cat'], drop_first=True)
```

```python
test=feat_eng(test)
test = pd.get_dummies(test, columns=['age_category', 'blood_pressure_category', 'blood_sugar_cat'], drop_first=True)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report


x_train = df.drop(columns=['smoking'])
y_train = df['smoking']

model = XGBClassifier()
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(test)
```

```python
sample_sub['smoking']=y_pred
sample_sub.to_csv('submission.csv',index=False)
sample_sub.head()
```
